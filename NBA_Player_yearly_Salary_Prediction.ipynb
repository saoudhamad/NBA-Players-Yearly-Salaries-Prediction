```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
stats_url="https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
salaries_url='https://www.basketball-reference.com/contracts/players.html'
```

```python
stats = pd.read_html(stats_url)
contracts = pd.read_html(salaries_url)
```

```python
stats = pd.concat(stats)
contracts = pd.concat(contracts)
```

```python
stats[26:32]
```

```python
#Some rows like 28 are errors we need to deal with them
stats = stats[stats.Rk != 'Rk']
```

```python
stats.columns
```

```python
#We only keep the relevant columns
stats_short = stats[['Player','Pos', 'Age', 'G', 'GS', 'MP', 'TRB', 'AST', 'BLK', 'TOV', 'PTS']]
```

```python
#Clean contracts dataframe
contracts.head()
```

```python
contracts.columns
```

```python
#We have multiIndex columns, we need to drop the two first levels
contracts.columns=contracts.columns.droplevel(0)
```

```python
contracts.head()
```

```python
contracts[19:23]
```

```python
#We need to clean the rows as well
contracts = contracts[contracts.Rk != 'Rk']
```

```python
#We need to convert the salaries to float
contracts['2021-22'] = contracts['2021-22'].str.replace(',', '')
contracts['2021-22'] = contracts['2021-22'].str.replace('$', '')
```

```python
contracts.head()
```

```python
#We keep only the relevant rows
contracts_short = contracts[['Player','2021-22']]
```

```python
#We have two clean dataframes let's combine them by using inner join
#Check common players in both datasets
contracts_short['Player'].isin(stats_short['Player']).value_counts()
```

```python
#Join the tables
full_data = pd.merge(stats_short,
                 contracts_short,
                 on='Player', 
                 how='inner')
```

```python
full_data.info()
```

```python
#drop rows with nas
full_data.dropna(inplace=True)
```

```python
#Change data type of columns
numeric = ['Age','G','GS','MP','TRB','AST','BLK','TOV','PTS','2021-22']
full_data[numeric] = full_data[numeric].apply(pd.to_numeric)
```

```python
#Assign numbers to the position column
full_data['Pos'].value_counts()
```

```python
d = {'SG': 2, 'PG': 1, 'C': 5, 'SF':3, 'PF':4,
     'PF-SF':4,'SF-PF':3, 'SF-SG':3, 'SG-SF':2, 'SG-PG':2, 'C-PF':5, 'PF-C':4, 'PG-SG':1 }
full_data['Pos'] = full_data['Pos'].map(d).fillna(full_data['Pos'])
```

```python
#Drop Player column
full_data.drop(['Player'], axis=1,inplace=True)
```

```python
#Create a new column which will be the target variable
full_data['target']=full_data['2021-22']/1000000
```

```python
full_data.head()
```

```python
full_data.info()
```

```python
"""
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
scaler = preprocessing.StandardScaler()
numeric = ['Age','G','GS','MP','TRB','AST','BLK','TOV','PTS','2021-22']
full_data[numeric] = scaler.fit_transform(full_data[numeric])
"""
```

```python
#split data train/test data with a 70/30 ratio
from sklearn.model_selection import train_test_split
X = full_data.drop(['2021-22','target'],axis=1)
y = full_data['target']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=10)
```

```python

#Scale the numerical columns
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
scaler = preprocessing.StandardScaler()
numeric = ['Age','G','GS','MP','TRB','AST','BLK','TOV','PTS']
X_train[numeric] = scaler.fit_transform(X_train[numeric])
X_test[numeric] = scaler.transform(X_test[numeric])

"""
y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)
y_train.head()
"""
```

```python
#Use Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
reg = RandomForestRegressor().fit(X_train,y_train)
y_pred = reg.predict(X_test)
mean_squared_error(y_test, y_pred)
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
y_pred = reg.predict(X_test)
import warnings
warnings.filterwarnings("ignore")
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
from sklearn.svm import SVR
reg = SVR().fit(X_train,y_train)
y_pred = reg.predict(X_test)
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
